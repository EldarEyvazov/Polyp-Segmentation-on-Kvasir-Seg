**Metrics visualizations**

```python
import pandas as pd
```

```python
ND_NA_train = pd.read_csv('models/DiceLoss_try_NonAug_NDS/log.csv')
ND_NA_test = pd.read_csv('models/DiceLoss_try_NonAug_NDS/test_log.csv')

ND_A_train = pd.read_csv('models/DiceLoss_try_Aug_NDS/log.csv')
ND_A_test = pd.read_csv('models/DiceLoss_try_Aug_NDS/test_log.csv')

D_NA_train = pd.read_csv('models/DiceLoss_try_NonAug/log.csv')
D_NA_test = pd.read_csv('models/DiceLoss_try_NonAug/test_log.csv')

D_A_train = pd.read_csv('models/DiceLoss_try_Aug/log.csv')
D_A_test = pd.read_csv('models/DiceLoss_try_Aug/test_log.csv')
```

```python
# Define a function to print metrics
def print_metrics(test_log):
    print('test_loss: %.4f - test_iou: %.4f - test_dice: %.4f - test_accuracy: %.4f'
          % (test_log['loss'], test_log['iou'], test_log['dice_coef'], test_log['accuracy']))

# Print metrics for each test log
print("Non-Augmented Attention U-Net:")
print_metrics(ND_NA_test)

print("\nAugmented Attention U-Net:")
print_metrics(ND_A_test)

print("\nNon-Augmented Deeper Attention U-Net:")
print_metrics(D_NA_test)

print("\nAugmented Deeper Attention U-Net:")
print_metrics(D_A_test)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Plot dice loss
plt.figure(figsize=(10, 6))
plt.plot(ND_A_train["epoch"],ND_A_train["val_iou"], linestyle = "--", label="IoU w/o DS", c="tab:orange")
plt.plot(ND_A_train["epoch"],ND_A_train["val_dice"], label="Dice w/o DS", c="tab:orange")
plt.plot(D_A_train["epoch"], D_A_train["val_iou"], linestyle = "--", c="tab:blue", label="IoU w/ DS")
plt.plot(D_A_train["epoch"], D_A_train["val_dice"], c="tab:blue", label="Dice w/ DS")
# plt.plot(ND_A_train["epoch"],ND_A_train["val_iou"], label="Training Dice Loss")
# plt.plot(ND_A_train["epoch"],ND_A_train["val_dice"], label="Training Dice Loss")
# plt.plot(D_A_train["epoch"], D_A_train["val_iou"], label="Training Dice Loss")
# plt.plot(D_A_train["epoch"], D_A_train["val_dice"], label="Training Dice Loss")
plt.xlabel("Epoch")
plt.ylabel("Metric")
plt.legend()
plt.grid(False)
plt.savefig("results/dsvnds.png")
plt.show()

```
